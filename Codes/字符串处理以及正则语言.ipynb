### 正则表达式


```python
import re
input = '自然语言处理很重要 。 12abc789'
```

```python
pattern = re.compile(r'.')#匹配所有字符
```

```python
re.findall(pattern,input)
```

```python
pattern = re.compile(r'[a-zA-Z]')#匹配所有英文字符
re.findall(pattern,input)
```

```python
pattern = re.compile(r'[^a-zA-Z]')#不匹配所有英文字符
re.findall(pattern,input)
```

```python
pattern = re.compile(r'\d') #匹配数字
re.findall(pattern,input)
```

```python
pattern = re.compile(r'\w')  #匹配字符和数字
re.findall(pattern,input)
```

```python
pattern = re.compile(r'\W')  #匹配非字母和数字
re.findall(pattern,input)
```

```python
pattern = re.compile(r'\s')   #匹配间隔符
re.findall(pattern,input)
```

```python
pattern = re.compile(r'\d*')  #  *表示0次或者多次匹配
re.findall(pattern,input)
```

```python
pattern = re.compile(r'\d+')  #  +表示一次或者多次匹配
re.findall(pattern,input)
```

```python
pattern = re.compile(r'\d?')  #  ?表示0或1次的匹配
re.findall(pattern,input)
```

```python
pattern = re.compile(r'\d{3}')#‘{m}’ 精确匹配 m 次
re.findall(pattern,input)
```

```python
pattern = re.compile(r'\d{1,3}')#{m,n}’ 匹配最少 m 次，最多 n 次。 (n>m)
re.findall(pattern,input)
```

# match 与 search 
match 从字符串的开头开始匹配，如果开头位置没有匹配成功，就算失败了；而 search 会跳过开头，继续向后寻找是否有匹配的字符串



```python
input2 = '自然语言123处理'
```

```python
pattern =  re.compile(r"\d")
e = re.match(pattern,input2)
e
```

```python
e = re.search(pattern,input2)
e
```

# 替换和修改   sub subn
sub 返回一个被替换的字符串

subn 返回一个元组，第一个元素是被替换的字符串，第二个元素是一个数字，表明产生了多少次替换

```python
pattern = re.compile(r'\d')
re.sub(pattern,'数字',input2)
```

```python
pattern = re.compile(r'\d')
re.subn(pattern,'数字',input2)
```

# 结合前面的

```python
input3 = '自然语言处理123机器学习456深度学习'
pattern = re.compile(r'\d+')
re.split(pattern,input3)
```

‘(?P…)’ 命名组

```python
input3 = '自然语言处理123机器学习456深度学习'
pattern = re.compile(r'(?P<dota>\d+)(?P<lol>\D+)')
m = re.search(pattern,input3)
m.groups('lol')
```

```python
input3 = '自然语言处理123机器学习456深度学习'
pattern = re.compile(r'(?P<lol>\D+)(?P<dota>\d+)')
m = re.search(pattern,input3)
m.groups()
```

```python
input3 = '自然语言处理123机器学习456深度学习'
pattern = re.compile(r'(?P<lol>\D+)(?P<dota>\d+)')
m = re.search(pattern,input3)
m.group('lol')
```

```python
# 筛选号码
input = 'number 338-343-220'
pattern = re.compile(r'(\d\d\d-\d\d\d-\d\d\d)')
m = re.search(pattern,input)

print(m.groups())
```
